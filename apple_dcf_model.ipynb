```python
import pandas as pd

revenue = 100000
growth_rate = 0.08
cost_ratio = 0.60

years = []
revenues = []
costs = []
profits = []

for year in range (1,6):
    revenue = revenue * (1 + growth_rate)
    cost = revenue * cost_ratio
    profit = revenue - cost

    years.append(year)
    revenues.append(revenue)
    costs.append(cost)
    profits.append(profit)

df = pd.DataFrame({
    "Year": years,
    "Revenue": revenues,
    "Costs": costs,
    "Profit": profits
})

df
```

## Building the DCF function

```python
def simple_dcf(ticker, fcf_margin=0.25, discount_rate=0.09, long_term_growth=0.025, forecast_years=5):
    company = yf.Ticker(ticker)
    revenue_row = company.financials.loc["Total Revenue"]
    revenue_billions = (revenue_row / 1e9).sort_index()

    growth_rates = revenue_billions.pct_change()
    avg_growth = growth_rates.dropna().mean()

    latest_revenue = revenue_billions.iloc[-1]
    rev = latest_revenue
    forecast_fcf = []

    for year in range(1, forecast_years + 1):
        rev = rev * (1 + avg_growth)
        fcf = rev * fcf_margin
        forecast_fcf.append(fcf)

    discounted_fcf = [fcf / (1 + discount_rate) ** (i + 1) for i, fcf in enumerate(forecast_fcf)]

    terminal_value = forecast_fcf[-1] * (1 + long_term_growth) / (discount_rate - long_term_growth)
    discounted_terminal_value = terminal_value / (1 + discount_rate) ** forecast_years

    enterprise_value = sum(discounted_fcf) + discounted_terminal_value

    print(f"{ticker} — Avg historical growth: {avg_growth:.2%}")
    print(f"{ticker} — Estimated Enterprise Value: ${enterprise_value:,.1f}B")

    return enterprise_value
```

## Sensitivity Analysis

```python
discount_rates_to_test = [0.07, 0.08, 0.09, 0.10, 0.11]

results = []
for dr in discount_rates_to_test:
    ev = simple_dcf ("AAPL", discount_rate=dr)
    results.append({"Discount Rate": dr, "Enterprise Value ($B)": ev})

sensitivity_df = pd.DataFrame(results)
sensitivity_df
```

```python
discount_rates_to_test = [0.08, 0.09, 0.10]
fcf_margins_to_test = [0.20, 0.25, 0.30]

results = []
for dr in discount_rates_to_test:
    for margin in fcf_margins_to_test:
        ev = simple_dcf("AAPL", discount_rate=dr, fcf_margin=margin)
        results.append({
            "Discount Rate": dr,
            "FCF Margin": margin,
            "Enterprise Value ($B)": ev
        })

sensitivity_df = pd.DataFrame(results)
sensitivity_df
```

## Visualizing the sensitivity grid

```python
import matplotlib.pyplot as plt
import numpy as np

pivot = sensitivity_df.pivot(index="Discount Rate", columns="FCF Margin", values="Enterprise Value ($B)")

fig, ax = plt.subplots()
im = ax.imshow(pivot.values, cmap="RdYlGn")

ax.set_xticks(range(len(pivot.columns)))
ax.set_xticklabels([f"{c:.0%}" for c in pivot.columns])
ax.set_yticks(range(len(pivot.index)))
ax.set_yticklabels([f"{r:.0%}" for r in pivot.index])

ax.set_xlabel("FCF Margin")
ax.set_ylabel("Discount Rate")
ax.set_title("Apple Enterprise Value Sensitivity ($B)")

for i in range(len(pivot.index)):
    for j in range(len(pivot.columns)):
        ax.text(j, i, f"{pivot.values[i, j]:,.0f}", ha="center", va="center", color="black")

plt.colorbar(im, label="Enterprise Value ($B)")
plt.show()
```
